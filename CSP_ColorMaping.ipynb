```python
# Territories:
# WA -> Western Australia
# NT -> Northern Territory
# SA -> South Australia
# Q -> Queensland
# NS -> New South Wales
# V -> Victoria
# T -> Tasmania


#Defining Colors
colors = ['Red', 'Green', 'Blue'] 

#Defining territories
states = ['WA', 'NT', 'Q', 'NSW' , 'V' , 'SA' , 'T']

#Defining Neighbors (dict datatype)
neighbors = {}

neighbors['WA'] = ['NT', 'SA']

neighbors['NT'] = ['WA', 'SA','Q']

neighbors['Q'] = [ 'NT', 'SA', 'NSW']

neighbors['NSW'] = ['V', 'SA' , 'Q']

neighbors['V'] = ['NSW', 'SA']

neighbors['SA'] = ['WA' ,'NT','Q' , 'NSW','V']

neighbors['T'] = ['']


#Defining territory color [initial colors of state] (dict datatype)
colors_of_states = {}

#Check color of neighbor territories 
def checkColor(state, color):
   for neighbor in neighbors.get(state):
        color_of_neighbor = colors_of_states.get(neighbor)
        if color_of_neighbor==color:
           return False
   return True

#Get individual color for neighbor territories 
def getColor(state):
  for color in colors:

     if checkColor(state, color): 
       return color

#Print the output (dict datatype)

for state in states:
    colors_of_states [state] = getColor(state) 

print (colors_of_states)
```
